```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 model (Nano version)
model = YOLO("yolov8n.pt")  

# Train the model
results = model.train(
    data=r"C:\Users\14252\OneDrive\Documents\School\wsu\2025Fall\421\fixformat\src\data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    patience=10
)

print("done")
```
